```python
import pandas # to read the data 
import nltk # for pre_processing and stopwords
import gensim # for Word2Vec Implementation
```

```python
# Getting the data from the dataset
review = pandas.read_excel('reviews.xlsx')
```

```python
review.head(5)
```

```python
# getting all the values to a list for easy manupulation
review_list = review['reviews.text'].values
```

```python
review_list[:2]
```

```python
# getting the tokens for each row of values
pre_processed_review =  [gensim.utils.simple_preprocess(i) for i in review_list]
```

```python
pre_processed_review[:5]
```

```python
#building the model for corpus, vector size is 150  and consider the word with count more than 2
model_gensi  =gensim.models.Word2Vec(pre_processed_review, size = 150, min_count=2)
```

```python
# Training  the model for 100 epochs
model_gensi.train(pre_processed_review,total_examples=len(pre_processed_review), epochs=100)
```

```python
# Getting word similarities for word good as it was amazon review you will see word as loud and great 
w1 = "good"
model_gensi.wv.most_similar(positive=w1)
```
